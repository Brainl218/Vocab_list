```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import os

DRIVER_PATH = './chromedriver.exe'
driver = webdriver.Chrome(executable_path=DRIVER_PATH)
```

```python
def vocab_scraping(url):
    driver.get(url)
    
    vocab_list = []
    all_uls = driver.find_elements_by_xpath('//div[@class="co"]')
    for ul in all_uls:
        all_vocabs = ul.find_elements_by_xpath('.//li')
        for vocab in all_vocabs:
            vocab_list.append(vocab.text)
    
    print(f"{len(vocab_list)} vocab found.")
    df = pd.DataFrame(vocab_list, columns=["vocab"])
    
    if not os.path.isdir("./vocab"):
        os.mkdir("./vocab")
    
    df.to_csv("./vocab/" + url[url.find("words.php?f=")+len("words.php?f="):] + ".csv", index=False)
```

```python
URL = "https://www.manythings.org/vocabulary/lists/c/"
    
r = requests.get(URL)

if r.status_code == requests.codes.ok:
    soup = BeautifulSoup(r.text, 'html.parser')
    
a_tags = soup.find_all("a")
for tag in a_tags:
    href = tag.get("href")
    if href.startswith("words"):
        v_url = URL + href
        print(v_url)
        vocab_scraping(v_url)
```
